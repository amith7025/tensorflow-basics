```python
import tensorflow as tf
print(tf.__version__)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
X = np.arange(-7.0,500.0,3)
y = np.arange(3.0,510.0,3.0)
X = X.reshape(-1, 1)  # Assuming X is a 1D array
y = y.reshape(-1, 1)
X,y,X.shape,y.shape
```

```python
plt.scatter(X,y)
```

```python
y == X + 10
```

```python
X = tf.constant(X)
y = tf.constant(y)
X,y
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),metrics=["mae"])

model.fit(X,y,epochs=1200)
```

```python
model.predict([100.0]).item()
```

```python
model.evaluate(X,y)
```
